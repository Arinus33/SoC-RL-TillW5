```python
# Create a function that grabs the email website domain
def get_domain(email_addr):
    return email_addr.split('@')[1]

get_domain('user@domain.com')
```

```python
# Create a basic function that returns True if the word 'dog' is contained in the input string.
def find_dog(sentence):
    return 'dog' in sentence.lower().split()

find_dog('Is there a dog here?')
```

```python
# Create a function that counts the number of times the word "dog" occurs in a string.
def count_dog(sentence):
    return sentence.lower().split().count('dog')

count_dog('This dog runs faster than the other dog dude!')
```

```python
# Use lambda expressions and the filter() function to filter out words from a list that don't start with the letter 's'.
seq = ['soup','dog','salad','cat','great']
list(filter(lambda word: word[0].lower() == 's', seq))
```

```python
# Final Problem
def caught_speeding(speed, is_birthday):
    shift = 5 if is_birthday else 0
    if speed <= 60 + shift:
        return 'No Ticket'
    elif 61 + shift <= speed <= 80 + shift:
        return 'Small Ticket'
    else:
        return 'Big Ticket'

caught_speeding(81,True)
```

```python
caught_speeding(81,False)
```
