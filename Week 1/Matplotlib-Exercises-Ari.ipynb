```python
import numpy as np
```

```python
m = np.linspace(0,10,11)
print(f"The array m should look like this: \n\n{m}")
```

```python
c = 3 * 10**8 # Speed of Light
```

```python
E = m*c**2
```

```python
print(f"The array E should look like this: \n\n {E}")
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(m, E, color='red', linewidth=5)
plt.title('E=mc^2')
plt.xlabel('Mass in Grams')
plt.ylabel('Energy in Joules')
plt.xlim([0, 10])
plt.show()
```

```python
plt.plot(m, E, color='red', linewidth=5)
plt.title('E=mc^2')
plt.xlabel('Mass in Grams')
plt.ylabel('Energy in Joules')
plt.xlim([0, 10])
plt.yscale('log')
plt.grid(axis='y', which='both')
plt.show()
```

```python
labels = ['1 Mo','3 Mo','6 Mo','1 Yr','2 Yr','3 Yr','5 Yr','7 Yr','10 Yr','20 Yr','30 Yr']
july16_2007 =[4.75,4.98,5.08,5.01,4.89,4.89,4.95,4.99,5.05,5.21,5.14]
july16_2020 = [0.12,0.11,0.13,0.14,0.16,0.17,0.28,0.46,0.62,1.09,1.31]
```

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(labels, july16_2007, label='july16_2007')
ax.plot(labels, july16_2020, label='july16_2020')
ax.legend()
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(labels, july16_2007, label='july16_2007')
ax.plot(labels, july16_2020, label='july16_2020')
ax.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))
plt.show()
```

```python
fig, axis1 = plt.subplots(figsize=(12, 8))
axis1.plot(labels, july16_2007, lw=2, color="blue")
axis1.set_ylabel("2007", fontsize=18, color="blue")
axis1.spines['left'].set_color('blue')
axis1.spines['left'].set_linewidth(4)
axis1.tick_params(axis='y', colors='blue', labelsize=15)
axis2 = axis1.twinx()
axis2.plot(labels, july16_2020, lw=2, color="red")
axis2.set_ylabel("2020", fontsize=18, color="red")
axis2.spines['right'].set_color('red')
axis2.spines['right'].set_linewidth(4)
axis2.tick_params(axis='y', colors='red', labelsize=15)
axis1.set_title("July 16th Yield Curves")
plt.show()
```
